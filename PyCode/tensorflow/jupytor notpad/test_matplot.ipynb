```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
"""
将2条线段放到一张图里
模拟x,y数据:
1. y1 = x^2 - 0.5 + noise(正太分布)
2. y2 = 2*x + 1
"""
# 返回等差数列
x = np.linspace(-1, 1, 300)
noise = np.random.normal(0, 0.05, x.shape)  # 正态分布:均值、标准差、shape
y1 = np.square(x) - 0.5 + noise
y2 = 2* x + 1


"""
将y1,y2放到一张图里
"""
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2, color='red', linestyle='--')
plt.show()

## gca
```

```python
"""
设置坐标轴
"""
x = np.linspace(-1, 1, 300)[:, np.newaxis]
noise = np.random.normal(0, 0.05, x.shape)  # 正态分布:均值、标准差、shape
y1 = np.square(x) - 0.5 + noise
y2 = 2* x + 1

plt.figure()
plt.plot(x, y1)

ax = plt.gca()  #get current axis
ax.spines['right'].set_color('none')  # 无色: 右边和顶部边框
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')  # 设置x坐标轴为下边框
ax.yaxis.set_ticks_position('left')  # 设置y坐标轴为左边框
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))


plt.show()
```
